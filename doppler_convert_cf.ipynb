### UMO to CF1.6 DSG Doppler timeseriesProfile

```python
#Convert UMAINE existing time series NetCDF doppler file in 4D (T,Z,Y,X) "grid" format to CF-1.6 DSG format
import os
import re
import netCDF4
import numpy as numpy
```

```python
buoy = 'B01'
# Get Doppler example.
file = 'http://www.neracoos.org/thredds/dodsC/UMO/All/B0125/realtime/B0125.doppler.realtime.nc'
ofile = 'DSG_' +  os.path.basename(file)
```

```python
nci = netCDF4.Dataset(file)

depth_dims = [d for d in nci.dimensions if re.search(r"depth", d) ]
print "DIMENSIONS:", nci.dimensions

the_dd = depth_dims[0]
# dopper has an rdi_depth and depth
the_dd = 'depth'

vars=nci.variables.keys()

coord_vars = ['time', 'lat', 'lon' ] + depth_dims

#data_vars = [var for var in vars if var not in coord_vars]
# To simplify for doppler, just get current_v and current_u
if re.search(r"doppler", file):
  data_vars = ['current_u', 'current_v']


nt = len(nci.dimensions['time'])
nz = len(nci.dimensions[the_dd])

print "Time Size,Depth Size",nt,nz
```

```python
# ## Write a sample "CF-1.6, timeSeriesProfile" File

title = buoy + ' '
title +=  'Buoy Sensor'

# create dimensions
nco = netCDF4.Dataset(ofile,'w',clobber=True)
nco.createDimension('time',nt)

if nz > 1:
  nco.createDimension('depth',nz)

station_name_strlen=5
nco.createDimension('name_strlen',station_name_strlen)

# create coordinate variables
time_v = nco.createVariable('time', 'f8', ('time'))
lon_v = nco.createVariable('lon','f4')
lat_v = nco.createVariable('lat','f4')
if nz > 1:
    depth_v = nco.createVariable('depth','f4',dimensions='depth')
else:
    depth_v = nco.createVariable('depth','f4')

station_v = nco.createVariable('station','S1',('name_strlen'))
```

```python
# write global attributes
g_attdict = nci.__dict__
g_attdict['Conventions'] = 'CF-1.6'
if nz>1:
    g_attdict['featureType'] = 'timeSeriesProfile'
else:
    g_attdict['featureType'] = 'timeSeries'

station_id = buoy

keywords = 'Oceans > Ocean Temperature > Water Temperature,Oceans'

g_attdict['station_name'] = g_attdict['mooring_site_id']
g_attdict['cdm_data_type'] = "Station"

g_attdict['publisher_name'] =  'Northeastern Regional Association of Coastal and Ocean Observing Systems (NERACOOS)'
g_attdict['publisher_email'] = 'info@neracoos.org'
g_attdict['publisher_phone'] = '(603) 319 1785'
g_attdict['publisher_url'] = 'http://www.neracoos.org/'
g_attdict['projejct'] = 'NERACOOS'


g_attdict['standard_name_vocabulary'] = 'CF-1.6'
g_attdict['summary'] = "NERACOOS UMaine Buoy Array"
g_attdict['naming_authority'] = 'edu.maine'
g_attdict['keywords']= keywords
g_attdict['keywords_vocabulary']='GCMD Science Keywords'
g_attdict['id'] = station_id
g_attdict['publisher'] = g_attdict['institution']
g_attdict['title'] = g_attdict['institution'] + ' ' + title
g_attdict['creator_email'] = g_attdict['contact']
g_attdict['creator_name'] =  'Neal Pettigrew'
g_attdict['creator_url'] = g_attdict['institution_url']
nco.setncatts(g_attdict) 
```

```python
# write station variable
station_v.name = station_id
station_v.long_name = station_id
station_v.short_name = station_id

station_v.cf_role = 'timeseries_id'
station_v.standard_name = 'station_name'
data = numpy.empty((1,),'S'+repr(station_name_strlen))
data[0] = station_id
station_v[:] = netCDF4.stringtochar(data)

# write time variable
time_v.units = 'days since 1858-11-17 00:00:00 +0:00'
time_v.standard_name = 'time'
time_v.calendar = 'gregorian'
time_v[:] = nci.variables['time'][:]

# write lon variable
lon_v.units = 'degree_east'
lon_v.standard_name = 'longitude'
lon_v[:] = nci.variables['lon'][:]

# write lat variable
lat_v.units = 'degree_north'
lat_v.standard_name = 'latitude'
lat_v[:] = nci.variables['lat'][:]

# write depth variable
depth_v.units = 'm'
depth_v.standard_name = 'depth'   
depth_v.positive = 'down'
depth_v.axis = 'Z'

depth_v[:] = nci.variables[the_dd][:]
```

```python
# create the data variables
for varname in data_vars:
    ncvar = nci.variables[varname]
    # if time series variable
    if numpy.size(ncvar) == nt:
        var = nco.createVariable(varname,ncvar.dtype,('time'))
    elif numpy.size(ncvar) == nz:
        var = nco.createVariable(varname,ncvar.dtype,('depth'))
    else:
        var = nco.createVariable(varname,ncvar.dtype,('time','depth'))

#   load old variable attributes and modify if necessary 
    attdict = ncvar.__dict__
    attdict['coordinates'] = 'time lon lat depth'

    # write variable attributes
    var.setncatts(attdict) 
    # write the data
    # current_u and current_v show up as Grid
    var[:] = ncvar[:]
```

```python
nci.close()
nco.close()

print "Created", ofile
```
